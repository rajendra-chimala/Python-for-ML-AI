# Filtering and Sorting

```python
import pandas as pd
```

```python
df = pd.read_csv('students_dataset.csv')
```

```python
df
```

```python
k = (df["Last_Name"]=="Kohli")
df.loc[k]
```

```python
df.loc[k,["First_Name","Last_Name","Roll_Number"]]
```

```python
k = (df["Last_Name"]!="Kohli")
df.loc[k]
```

```python
df.loc[~k,["First_Name","Last_Name","Roll_Number"]]
```

```python
k = (df["Last_Name"] == "Kohli") | (df["Last_Name"] == "Gupta")
df.loc[k]
```

```python
k = df["Last_Name"].isin(["Kohli", "Gupta"])
df.loc[k]
```

#**Sorting**

```python
df.sort_values(by="First_Name")
```

```python
df.sort_values(by="First_Name", ascending=False)
```

```python
df.sort_index(inplace=True)
```
