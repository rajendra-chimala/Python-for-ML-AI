#Loc and iLoc
The 'loc' function is mainly used when we want to select rows and columns based on their labels

Whereas 'iloc' is used when choosing rows and columns based on specific positions

```python
df1.loc[[0,1,2],['Hobbyist']]
```

```python
df1.loc[[0,1,2,3],['Hobbyist','OpenSourcer']]
```

```python
df1.loc[[0,1,2,3],['Hobbyist','OpenSourcer','Country']]
```

```python
df1.loc[0:3,'Hobbyist':'Country']
```

```python
df1.iloc[[0,1,2],[0,1,2,3]]
```

```python
df1.iloc[0:4,0:5]
```
