```python
def iqr_bounds(data,column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3-Q1
    lower = Q1-1.5*IQR
    upper = Q3+1.5*IQR
    return lower,upper 

def remove_outliers(data,column):
    lower,upper = iqr_bounds(data,column)
    filt = (data[column]>=lower) & (data[column]<=upper)
    return data.loc[filt]

# def remove_outliers(data,column):
#     lower,upper = iqr_bounds(data,column)
#     filt = (data[column]>=lower) & (data[column]<=upper)
#     return data.loc[filt]
```

```python
def iqr_bounds(data,column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3-Q1
    lower = Q1-1.5*IQR
    upper = Q3+1.5*IQR
    return lower,upper 

def remove_outliers(data,column):
    # print(f"Data : {data},Column : {column}")
    lower,upper = iqr_bounds(data,column)
    filt = (data[column]>=lower) & (data[column]<=upper)
    return data.loc[filt]

df_cleaned = (
    df.pipe(remove_outliers,column="Salary").pipe(remove_outliers,column="Age")
)

# df
# df_cleaned


df_cleaned.to_csv('./sample_data/cleaned_outlier_dataset.csv',index=False)
```
